```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

X = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
Y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

# Xp = np.sort(np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110]))
Xp = np.arange(25, 200, 1)
Yp = 1000 - (17000/(Xp-5))
Yr = 2.6205389*Xp+444.17736

fig, (ax1) = plt.subplots(
    nrows=1, ncols=1,
    figsize=(8, 4)
)

ax1.scatter(x=X, y=Y, c='purple')
ax1.set_title('Корреляция зарплаты и скорингововго коэффициента')
ax1.set_xlabel('Заработная плата')
ax1.set_ylabel('Кредитный скоринг')
ax1.set_ylim([0, 1000])
ax1.plot(Xp, Yp, c='lightgrey')
ax1.plot(Xp, Yr, c='red', linewidth = 4)
```
