```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.linspace(-10, 10, 42)

plt.axis([-10, 10, -1.5, 1.5])

plt.title('Cosines')
plt.xlabel('x')
plt.ylabel('F(x)')

plt.plot(x, np.cos(x), marker="o", label = 'F(x) = cos(x)')
plt.plot(x, np.cos(0.5*x), marker="o", label = 'F(x) = cos(x/2)')

plt.legend(loc = 'upper right')
```
