```python
import numpy as np
from matplotlib import pyplot as plt

a = np.linspace(0.01, 0.02, 100)

n = 3
x1 = n*np.pi / a
n = 2
x2 = n*np.pi / a
n = 1
x3 = n*np.pi / a

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(a, x1)
ax.plot(a, x2)
ax.plot(a, x3)

plt.xlabel("a")
plt.ylabel("x")
ax.grid()
plt.ylim((100, 500))

plt.show()
```
