```python
# Напишите код на Python, реализующий построение графиков:
# 1.окружности,
# 2.эллипса,
# 3.гиперболы.

import numpy as np
from matplotlib import pyplot as plt
from math import pi, cos, sin

u=1.       #x-position of the center
v=0.5      #y-position of the center
r = 1.5

t = np.linspace(0, 2*pi, 100)

fig, (ax_1, ax_2, ax_3) = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

Crl = np.array([r*np.cos(t) , r*np.sin(t)]) 

ax_1.plot( u+Crl[0,:] , v+Crl[1,:], color = 'b' )     #initial circle
ax_1.set_aspect(1)

# Эллипс
a=2.       #radius on the x-axis
b=1.5      #radius on the y-axis

Ell = np.array([a*np.cos(t) , b*np.sin(t)])  
     #u,v removed to keep the same center location

ax_2.plot( u+Ell[0,:] , v+Ell[1,:], color = 'b' )     #initial ellipse
ax_2.set_aspect(1)

# Гипербола
t = np.linspace(-10, 10, 100)
Hpb = np.array([t, 1/t]) 
ax_3.plot( u+Hpb[0,:50] , v+Hpb[1,:50], color = 'b')     #initial hyperbole
ax_3.plot( u+Hpb[0,50:] , v+Hpb[1,50:], color = 'b')     #initial hyperbole
ax_3.set_aspect(1)

plt.show()
```

```python
# 1.Напишите код, который будет переводить полярные координаты в декартовы.
# 2.Напишите код, который будет рисовать график окружности в полярных координатах.
# 3.Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

# Соединил все в одном скрипте. Есть 2 функции перевода декартовых координат в радиальные и обратно.
# Окружность нарисована и в радиальных и в декартовых координатах. Причем перевод координат был произведен 
# с помощью написанных функций
# Есть график прямой. Координаты которой из декартовых переведны в радиальные и отрисованы соответсвтующим обоазом. 

import numpy as np
from matplotlib import pyplot as plt


def convert_d_p(x, y):
    r = np.sqrt(x ** 2 + y ** 2)
    phi = np.arctan2(y, x)

    return r, phi


def convert_p_d(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)

    return x, y


fig = plt.figure(figsize=(30, 6))
ax_1 = plt.subplot(151, projection='polar')
ax_2 = plt.subplot(153)
ax_3 = plt.subplot(155, projection='polar')

# Круг в полярных координатах
phi = np.linspace(0, 2 * np.pi, 20)
r = 2
ax_1.plot(phi, [r] * len(phi))  # circle in polar coordinates
ax_1.set_aspect(1)

# Проверка, что полярные координаты правильно переводятся в декартовы
x, y = convert_p_d(r, phi)
ax_2.plot(x, y)  # circle in cartesian coordinates
ax_2.grid(True)
ax_2.set_aspect(1)

# Прямая, которая выводится на графике в полярных координатах
x = np.linspace(-4.5, 2, 20)
y = 2 * x + 3
r, phi = convert_d_p(x, y)
ax_3.plot(phi, r)  # line in polar coordinates
ax_3.set_aspect(1)

plt.show()
```

```python
# 1)Решите систему уравнений:
# y = x2 – 1
# exp(x) + x∙(1 – y) = 1

import numpy as np
from matplotlib import pyplot as plt

fig = plt.figure()
ax_1 = plt.subplot(111)

x1 = np.linspace(-5, 5, 100)
y1 = x1**2-1

ax_1.plot(x1, y1)

y2 = (np.exp(x1) + x1  - 1) / x1

ax_1.plot(x1, y2)
```

```python
# Из графика видно, что есть 3 точки пересечения. Их и нужно найти. укажем примерное их положение в функции fsolve
# (-2, 2), (2, 6), (4, 20)

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)
xs, ys = fsolve(equations, (-2, 2))
print(xs, ys)

xs, ys = fsolve(equations, (2, 6))
print(xs, ys)

xs, ys = fsolve(equations, (4, 20))
print(xs, ys)
```
